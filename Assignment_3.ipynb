```python
# Creating a list of books
library_books = []

# Adding books to the list
library_books.append("Harry Potter")
library_books.append("Atomic Habits")
library_books.append("The Alchemist")

# Printing the updated list
print("Books in Library:")
print(library_books)
```

```python
# Display books with index numbers
print("Library Book List:")

for index, book in enumerate(library_books):
    print(f"{index}: {book}")
```

```python
# Function to check if a book exists
def check_book_availability(book_name):
    if book_name in library_books:
        return f"{book_name} is available."
    else:
        return f"{book_name} is not available."

# Testing the function
print(check_book_availability("Atomic Habits"))
print(check_book_availability("Twilight"))
```

```python
with open("books.txt", "w") as file:
    file.write("Harry Potter\n")
    file.write("Atomic Habits\n")
    file.write("The Alchemist\n")
```

```python
# Function to read books from a text file
def read_books_from_file(filename):
    try:
        with open(filename, "r") as file:
            books = file.readlines()

        print("Books from file:")

        for book in books:
            print(book.strip())

    except FileNotFoundError:
        print("Error: File not found.")


        # Test function
read_books_from_file("books.txt")
```

```python
import pickle

# Save books using pickle
with open("books.pkl", "wb") as file:
    pickle.dump(library_books, file)

print("Books saved successfully.")
```

```python
# Function to load books using pickle
def load_books_pickle(filename):
    try:
        with open(filename, "rb") as file:
            books = pickle.load(file)

        print("Loaded Books:")
        print(books)

    except FileNotFoundError:
        print("Pickle file not found.")

# Test function
load_books_pickle("books.pkl")
```

```python
import sqlite3

# Function to add books to database
def add_books_to_database(book_name):

    try:
        conn = sqlite3.connect("library.db")
        cursor = conn.cursor()

        # Create table
        cursor.execute("""
        CREATE TABLE IF NOT EXISTS books (
            id INTEGER PRIMARY KEY AUTOINCREMENT,
            title TEXT
        )
        """)

        # Insert book
        cursor.execute(
            "INSERT INTO books (title) VALUES (?)",
            (book_name,)
        )

        conn.commit()

        print(f"{book_name} added to database successfully.")

    except sqlite3.Error as error:
        print("Database Error:", error)

    finally:
        conn.close()

# Test function
add_books_to_database("Rich Dad Poor Dad")
```

```python
# View books stored in database

conn = sqlite3.connect("library.db")
cursor = conn.cursor()

cursor.execute("SELECT * FROM books")

rows = cursor.fetchall()

print("Books in Database:")

for row in rows:
    print(row)

conn.close()
```

```python
def remove_book(book_name):

    if book_name in library_books:
        library_books.remove(book_name)
        print(f"{book_name} removed successfully.")
    else:
        print("Book not found.")

# Test
remove_book("Harry Potter")

print(library_books)
```
