```python
from transformers import AutoTokenizer, AutoModelForCausalLM

model_name = "gpt2"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)
```

```python
def generate_text(prompt, max_length=50, temperature=1.0, top_k=50, top_p=0.95):
    inputs = tokenizer(prompt, return_tensors="pt")

    output = model.generate(
        **inputs,
        max_length=max_length,
        temperature=temperature,
        top_k=top_k,
        top_p=top_p,
        do_sample=True
    )

    return tokenizer.decode(output[0], skip_special_tokens=True)
```

```python
print(generate_text("Once upon a time"))
```

```python
print(generate_text("Once upon a time", max_length=100))
```

```python
print(generate_text("Once upon a time", temperature=0.5))  # safer
print(generate_text("Once upon a time", temperature=1.2))  # more creative
```

```python
print(generate_text("Once upon a time", top_k=10))
print(generate_text("Once upon a time", top_p=0.8))
```

```python
print(generate_text("Write a motivational message for a single mom balancing school and life:"))
```
